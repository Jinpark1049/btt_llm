```python
import streamlit as st

# PDF 업로드
with st.sidebar:
    st.title('📁 BTT Report Extractor')

    uploaded_file = st.file_uploader("Upload a PDF file", type=["pdf"])

    # 새로운 PDF가 업로드되면 실행
    if uploaded_file is not None:
        st.session_state.pdf_path = uploaded_file  # 세션 상태에 저장
        st.sidebar.write("✅ New PDF uploaded.")

    # LLM 실행 버튼
    if st.button("Run LLM Parser"):
        if "pdf_path" in st.session_state:
            parser.run(st.session_state.pdf_path)  # PDF 처리 실행
            st.session_state.df_result = parser.result  # 새로운 결과 저장
            st.sidebar.success("✅ LLM Parser run successfully!")

# 결과 표시 (Editable)
if "df_result" in st.session_state and st.session_state.df_result is not None:
    st.subheader("📊 Extracted Data (Editable)")

    # 기존 데이터 유지 옵션 (선택 사항)
    keep_old_data = st.checkbox("Keep previous data", value=False)

    if keep_old_data and "old_df" in st.session_state:
        # 기존 데이터와 새로운 데이터를 합치기
        updated_df = st.session_state.old_df.append(st.session_state.df_result, ignore_index=True)
    else:
        updated_df = st.session_state.df_result

    # 수정 가능하게 표시
    edited_df = st.data_editor(updated_df, num_rows="dynamic")

    # 수정된 데이터 세션에 저장
    st.session_state.df_result = edited_df
    st.session_state.old_df = edited_df  # 기존 데이터 유지

    st.success("✅ Changes saved in session state!")

    # CSV 다운로드 버튼
    csv = edited_df.to_csv(index=False).encode('utf-8')
    st.download_button("📥 Download CSV", csv, "edited_data.csv", "text/csv")
```
